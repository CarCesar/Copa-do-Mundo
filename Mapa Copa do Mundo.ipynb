# Adicionar ISO

```python
ISO=pd.read_csv('https://raw.githubusercontent.com/CarCesar/COVID/main/ISO.csv')
```

```python
df4['teste']=1/df4['Posicao']
df5=df4.merge(ISO,on='alpha2')
df6 = df5.groupby(by=['Selecao','id']).min()
df6.reset_index(inplace=True)
```

# Trocar Nomes

```python
nomes = ['Campeão','Vice Campeão','Terceiro','Quarto','Quartas de Final','Oitavas de Final','Fase de Grupos']
antigos = sorted(df6.P.unique())
for a in range(len(nomes)):
    df6.P.replace(antigos[a],nomes[a],inplace=True)
```

```python
df6.P.replace('Terceiro','Semi-Final',inplace=True)
df6.P.replace('Quarto','Semi-Final',inplace=True)
```

```python
def fazconta(a1,a2,d1,d2,intervalos):
    a = a1*16+a2
    d = d1*16+d2
    diff = (a - d)/(intervalos-1)
    lista = []
    lista.append([a1,a2])
    for b in range(1,intervalos-1):
        c=round(a - b*diff)
        c1=c//16
        c2 = c%16
        lista.append([c1,c2])
    lista.append([d1,d2])
    return lista

def definir_dgd(cor1,cor2,intervalos):
    dic = {10:'a',11:'b',12:'c',13:'d',14:'e',15:'f'}
    dct = {v: k for k, v in dic.items()}
    
    def diz(v):
        if v > 9:
            return dic[v]
        else:
            return str(v)
    
    def rediz(v):
        if v in dct.keys():
            return dct[v]
        else:
            return int(v)
    red = fazconta(rediz(cor1[1]),rediz(cor1[2]),rediz(cor2[1]),rediz(cor2[2]),intervalos)
    green = fazconta(rediz(cor1[3]),rediz(cor1[4]),rediz(cor2[3]),rediz(cor2[4]),intervalos)
    blue = fazconta(rediz(cor1[5]),rediz(cor1[6]),rediz(cor2[5]),rediz(cor2[6]),intervalos)
    lista = []
    for a in range(intervalos):
        st='#'
        st=st+ diz(red[a][0])+diz(red[a][1])+diz(green[a][0])+diz(green[a][1])+diz(blue[a][0])+diz(blue[a][1])
        lista.append(st)
    return lista
```
