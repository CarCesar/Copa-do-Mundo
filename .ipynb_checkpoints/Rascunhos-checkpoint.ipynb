```python
input_dropdown = alt.binding_select(options=df.Continente.unique(), name='Country')
selection = alt.selection_single(fields=['Continente'])
sel = alt.selection_single(fields=['Selecao'])
sel1 = alt.selection_multi(fields=['Selecao'],empty='none')

line = alt.Chart(df).mark_line(point=True,line=True,color='gold').encode(
    x = alt.X("Ano:O"),# timeUnit="year", title="date"),
    y="Posicao:O",
    color=alt.Color('Selecao:N',scale=alt.Scale(range=['gold'])),
    opacity=alt.condition(sel1, alt.value(1),alt.value(0)),
    tooltip=['Selecao:N'],
    shape='Selecao:N',

).properties(
    width=1200,
    height=800,
)#.add_selection(selection)
```

```python
band0=alt.Chart(df).mark_image(size=500).encode(
    x = alt.X("Ano:O"),# timeUnit="year", title="date"),
    y="Posicao:O",
    #color=alt.condition(selection,alt.Color("Selecao:N",legend=None),alt.value('gray')),
    opacity=alt.condition(selection,alt.value(0.2),alt.value(0)),
    tooltip=['Selecao:N'],
    shape='Continente:N',
    url='flag',
    #width=alt.value(10)
).properties(
    width=1200,
    #height=800,
)

band0.add_selection(selection,sel,sel1)
```

```python
band=alt.Chart(df).mark_image(height=10).encode(
    x = alt.X("Ano:O"),# timeUnit="year", title="date"),
    y="Posicao:O",
    #color=alt.condition(selection,alt.Color("Selecao:N",legend=None),alt.value('gray')),
    opacity=alt.condition(sel,alt.value(1),alt.value(.05)),
    #width=alt.condition(sel,alt.value(25),alt.value(15)),
    tooltip=['Selecao:N'],
    shape='Continente:N',
    url='flag'
).properties(
    width=800,
    height=600,
)

band.add_selection(selection,sel,sel1)
```

```python
ISO=pd.read_csv('https://raw.githubusercontent.com/CarCesar/COVID/main/ISO.csv')
```

```python
ISO
```

```python
ISO['alpha3']=[a.upper() for a in ISO['alpha3']]
```

```python
df2=df.merge(ISO,on='alpha2')
```

```python
df2['teste']=1/df2['Posicao']
```

```python
df3 = df2.groupby(by=['Selecao','id']).min()
```

```python
df3
```

```python
df3.reset_index(inplace=True)
```

```python
df3
```

```python
#df3.sort_values('teste',ascending=False).head(60).reset_index()
```

```python
from vega_datasets import data

terra = alt.topo_feature(data.world_110m.url,'countries')
sphere = alt.sphere()
graticule = alt.graticule()
mar = alt.Chart(sphere).mark_geoshape(fill='lightgrey')#lightblue
linhas = alt.Chart(graticule).mark_geoshape(stroke='#444444', strokeWidth=0.5)
fundo = alt.Chart(terra).mark_geoshape(stroke="#444444", strokeWidth=0.15,fill='#444444')#lightgrey
```

```python
cor="goldred"
mapa = alt.Chart(terra
         ).mark_geoshape(stroke="black", strokeWidth=0.15).transform_lookup(
    lookup = 'id',from_=alt.LookupData(data=df3, key='id', fields=['Posicao','Selecao','teste'])
    ).encode(
    color = alt.Color('Posicao:Q', scale=alt.Scale(scheme=cor),sort='ascending',legend=alt.Legend(type='gradient',
                                                                                                  direction='horizontal',
                                                                                                  orient='bottom-right',
                                                                                                  fillColor='lightgrey',
                                                                                                  padding=9,
                                                                                                  gradientLength=380,
                                                                                                  offset=2,strokeColor='black')),
    #range=['gold','red']
    tooltip = [alt.Tooltip('Selecao:N',title = 'Country'),alt.Tooltip('Posicao:Q')]
    )

fig = (mar+fundo+linhas+mapa).properties(width=800, height=400,title='Copa do Mundo - Melhor posição de Cada Seleção'
                        ).project('equirectangular').configure_view(strokeWidth=0)
```
