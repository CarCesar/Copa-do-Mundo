```python
b = a[2].iloc[:,1]
d=[]
for c in b:
    d.append(c.split()[0])
a[2].iloc[:,1]=d
```

```python
a[2]
```

```python
b = a[2].iloc[:,3]
d=[]
e = []
for c in b:
    d.append(int(c.split()[0]))
    e.append(c.split(' (')[1].split(')')[0].split(', ')[-1])
a[2].iloc[:,3]=d
a[2]['last']=e
```

```python
a[2]
```

```python
b = a[2].iloc[:,4]
d=[]
e=[]
f=[]
for c in b:
    d.append(c.split()[0])
    if c.split(' (')[-1].split(')')[0].split(', ') == ['—']:
        e.append(0)
        f.append(0)
    else:
        e.append(len(c.split(' (')[-1].split(')')[0].split(', ')))
        f.append(c.split(' (')[-1].split(')')[0].split(', ')[-1])
a[2].iloc[:,4]=d
a[2]['best_times']=e
a[2]['best_last']=f
```

```python
a[2].columns = ['Team', 'Qualified', 'date',
       'appearances', 'bestperformance',
       'last', 'best_times','best_last']
```
